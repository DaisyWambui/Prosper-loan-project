## Investigation Overview
In this investigation I focused on EmploymentStatus,LoanStatus,CurrentlyInGroup and how the named variables affect EstimatedReturn,EstimatedLoss,LenderYield,BorrowerAPR,etc.


## Dataset Overview

> The dataset includes 81 columns.Some of the investigated columns include EmploymentStatus,LoanStatus,CurrentlyInGroup,EstimatedReturn,EstimatedLoss,LenderYield,BorrowerAPR to mention a few.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_df=pd.read_csv('prosperLoanData.csv')
```

## (Distribution of EmploymentStatus)



There are different EmploymentStatus:Employed,self-employed,fulltime,not available,other,part-time,not-employed and retired.
<br/>This visualization shows the distribution of employment status by count.

```python
plt.figure(figsize=[8,6])
sns.set(style='white')
sns.countplot(data=loan_df,x='EmploymentStatus',color='maroon',order=loan_df['EmploymentStatus'].value_counts().index)
plt.xticks(rotation=45)
plt.title('Distribution of EmploymentStatus')
```

## (Relationship between Borrower rate and Estimated loss)

> This visualization shows the relationship between borrower rate and estimated loss.
<br/>The correlation is positive

```python
plt.figure(figsize=(8,5))
sns.regplot(data=loan_df, x=loan_df['BorrowerRate'],y=loan_df['EstimatedLoss'])
plt.title('Relationship between Borrower rate and Estimated loss')
```

## (Relationship between Lender yield and borrower rate)
There's a high/positive correlation between Lenderyield and borrower rate.



```python
plt.figure(figsize = [8, 6])
plt.scatter(data = loan_df, x = 'LenderYield', y = 'BorrowerRate')
plt.title('Relationship between Lender yield and borrower rate')
plt.xlabel('LenderYield')
plt.ylabel('BorrowerRate')
```


    (Relationship between Lenderyield and EmploymentStatus)

>This visualization shows the Lenderyield for employed people is higher compared to the other statuses,followed by fulltime,self-employed then not available.

```python
g=sns.FacetGrid(data=loan_df,col='EmploymentStatus')
g.map(plt.hist,'LenderYield',bins=15)
```


(Does being a home owner affect Estimatedreturn and Estimatedloss?)

>Being a borrower who is a home owner or not does not affect estimatedreturn and estimated loss.

```python
g=sns.FacetGrid(data=loan_df,col='IsBorrowerHomeowner')
g.map(sns.regplot,'EstimatedLoss','EstimatedReturn',color='orange')
```
